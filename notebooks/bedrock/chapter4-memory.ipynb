## 学習ポイント
- メモリなし: 前の記事を忘れて一から生成
- メモリあり: 前の記事を覚えていて、差分だけで修正できる
- `thread`と`resource`による会話の識別

## ポイント解説

`Memory` クラスと `thread` / `resource` で会話履歴を管理します。

```typescript
import { Memory } from "@mastra/memory";
import { LibSQLStore } from "@mastra/libsql";

const memory = new Memory({
  storage: new LibSQLStore({ url: ":memory:" }),
  options: { lastMessages: 10 },
});

const agent = new Agent({ memory, /* ... */ });

// memory: { thread, resource } で会話を識別
await agent.generate("タイトル候補を3つ提案して", {
  memory: { thread: "session-1", resource: "user-1" },
});

// 同じ thread → 前の提案を覚えている
await agent.generate("2番目に決めた。構成案を出して", {
  memory: { thread: "session-1", resource: "user-1" },
});
```

メモリなしだと「2番目に決めた」が何を指すか不明。メモリありなら前の提案を覚えているので「2番目」が通じます。
