# Chapter 2: タスク分割とワークフロー設計 【Bedrock版】

このNotebookでは、タスクを分割してワークフローを設計する方法を学びます。

**使用モデル**: Amazon Nova Lite

## 学習ポイント
- `research → outline → write → review` のワークフロー設計
- **分割の本質的な難しさ**: 「どこで切るか」より「何を渡すか（スキーマ設計）」が重要
- 最初のステップ（research）の品質が、後続すべてに影響する問題

## ワークフローの流れ

```
researchStep (リサーチ)  → { keyPoints[], targetAudience, tone }
  ↓
outlineStep (構成設計)   → { sections[], targetAudience, tone }
  ↓
writeStep (執筆)        → { draft, targetAudience }
  ↓
reviewStep (レビュー)    → { revisions[], article }
```

## Chapter 2の実行

````python
import subprocess, os

def run(cmd):
    """UTF-8安全なコマンド実行（%%bashの代替）"""
    result = subprocess.run(cmd, shell=True, cwd=os.path.join(os.getcwd(), '..', '..'),
                            capture_output=True, text=True)
    if result.stdout: print(result.stdout, end='')
    if result.returncode != 0 and result.stderr: print(result.stderr, end='')
````
