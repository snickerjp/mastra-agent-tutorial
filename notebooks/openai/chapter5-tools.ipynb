## ポイント解説

`createTool` でエージェントが使える「アクション」を定義します。

```typescript
// ツールを定義
const searchTopic = createTool({
  id: "search-topic",
  description: "トピックに関する最新情報を検索する",  // ← LLM がこれを見て呼ぶか判断
  inputSchema: z.object({
    query: z.string().describe("検索クエリ"),  // ← LLM が引数を自律的に決める
  }),
  execute: async ({ query }) => { /* 検索実行 */ },
});

// Agent にツールを渡すだけ
const agent = new Agent({
  tools: { getCurrentDate, searchTopic },  // ← LLM がいつ呼ぶかを自律的に判断
  // ...
});
```
