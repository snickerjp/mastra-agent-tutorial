# Chapter 3: プロンプトの品質を測る 【OpenAI版】

このNotebookでは、3つのプロンプトパターンを同じ評価基準で比較し、スコアの違いを確認します。

## 学習ポイント
- プロンプトの品質がアウトプットに与える影響を数値で比較
- 曖昧な指示と明確な指示の違い
- 評価スコアによる客観的な品質測定

## 3つのパターン

| パターン | instructions | リクエスト | 期待スコア |
|---|---|---|---|
| A（最悪）| 「記事を書いてください」 | 「TypeScriptについて」 | 低 |
| B（中程度）| 詳しい役割設定あり | トピックのみ | 中 |
| C（最良）| 詳しい役割設定あり | 対象読者・構成・トーン指定あり | 高 |

## ポイント解説

`createScorer` で評価関数を定義し、エージェントの出力を数値で採点します。

```typescript
// 評価関数を定義
const contentQualityScorer = createScorer<Input, Output>({
  name: "コンテンツ品質",
  description: "構成・深さ・具体性・読者への適切さの総合評価",
  execute: async ({ input, output }) => {
    // LLM に採点させて 0〜1 のスコアを返す
    return { score: 0.85 };
  },
});

// 3 パターンを同じ評価基準で比較
// A: 曖昧な指示 → 低スコア
// B: 詳しい役割設定 → 中スコア
// C: 役割 + 読者・構成・トーン指定 → 高スコア
```

「曖昧な指示でもエラーにならない」のがエージェント開発の難しさ。スコアで数値化することで、品質の差を客観的に確認できます。

代わりに「低品質なアウトプット」が返ってきます。

```typescript
// ❌ 曖昧な指示 → エラーにならないが品質が低い

// ✅ 明確な指示 → 安定した高品質なアウトプット
```
