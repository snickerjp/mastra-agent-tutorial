# Chapter 2: タスク分割とワークフロー設計 【OpenAI版】

このNotebookでは、タスクを分割してワークフローを設計する方法を学びます。

## 学習ポイント
- `research → outline → write → review` のワークフロー設計
- **分割の本質的な難しさ**: 「どこで切るか」より「何を渡すか（スキーマ設計）」が重要
- 最初のステップ（research）の品質が、後続すべてに影響する問題

## ワークフローの流れ

```
researchStep (リサーチ)  → { keyPoints[], targetAudience, tone }
  ↓
outlineStep (構成設計)   → { sections[], targetAudience, tone }
  ↓
writeStep (執筆)        → { draft, targetAudience }
  ↓
reviewStep (レビュー)    → { revisions[], article }
```

## ポイント解説

タスクを 4 ステップに分割し、各ステップの出力スキーマを明確に定義します。

```typescript
// 各ステップの「出力」が次のステップの「入力」になる
const researchStep = createStep({
  id: "research",
  outputSchema: z.object({
    keyPoints: z.array(z.string()),   // ← 次のステップが必要とする情報を明示
    targetAudience: z.string(),
    tone: z.string(),
  }),
  // ...
});

// ワークフローでステップを連結
const workflow = createWorkflow({ steps: [researchStep, outlineStep, writeStep, reviewStep] });
```

本質的な難しさ: 「どこで切るか」より「次のステップに何を渡すか」の設計が重要です。最初の research ステップの出力スキーマが不十分だと、後続の outline → write → review すべてが影響を受けます。
