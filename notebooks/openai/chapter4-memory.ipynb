# Chapter 4: メモリで改善を重ねる 【OpenAI版】

このNotebookでは、メモリ機能を使って会話履歴を保持し、差分修正を実現する方法を学びます。

## 学習ポイント
- メモリなし: 前の記事を忘れて一から生成
- メモリあり: 前の記事を覚えていて、差分だけで修正できる
- `thread`と`resource`による会話の識別

## メモリの仕組み

```typescript
// thread と resource で会話を識別
await agent.generate("初稿を書いて", {
  memory: { thread: "article-session-1", resource: "user-1" }
});

// 同じ thread なので前の記事を覚えている
await agent.generate("もっと初心者向けに書き直して", {
  memory: { thread: "article-session-1", resource: "user-1" }
});
```
