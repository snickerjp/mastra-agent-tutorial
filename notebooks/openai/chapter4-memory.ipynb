# Chapter 4: メモリで改善を重ねる 【OpenAI版】

このNotebookでは、メモリ機能を使って会話履歴を保持し、差分修正を実現する方法を学びます。

## 学習ポイント
- メモリなし: 前の記事を忘れて一から生成
- メモリあり: 前の記事を覚えていて、差分だけで修正できる
- `thread`と`resource`による会話の識別

## ポイント解説

`Memory` クラスと `thread` / `resource` で会話履歴を管理します。

```typescript
import { Memory } from "@mastra/memory";
import { LibSQLStore } from "@mastra/libsql";

// インメモリ SQLite でストレージを作成
const memory = new Memory({
  storage: new LibSQLStore({ url: "file::memory:" }),
});

const agent = new Agent({ memory, /* ... */ });

// thread と resource で会話を識別
await agent.generate("初稿を書いて", {
  threadId: "article-session-1",   // ← 同じ thread なら前の記事を覚えている
  resourceId: "user-1",
});

await agent.generate("もっと初心者向けに書き直して", {
  threadId: "article-session-1",   // ← 同じ thread なので差分修正できる
  resourceId: "user-1",
});
```

メモリなしだと「もっと初心者向けに」と言っても前の記事を忘れて一から生成します。メモリありなら、前の記事を覚えていて差分だけ修正できます。
