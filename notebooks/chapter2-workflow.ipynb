# Chapter 2: タスク分割とワークフロー設計

このNotebookでは、タスクを分割してワークフローを設計する方法を学びます。

## 学習ポイント
- `research → outline → write → review` のワークフロー設計
- **分割の本質的な難しさ**: 「どこで切るか」より「何を渡すか（スキーマ設計）」が重要
- 最初のステップ（research）の品質が、後続すべてに影響する問題

## ワークフローの流れ

```
researchStep (リサーチ)  → { keyPoints[], targetAudience, tone }
  ↓
outlineStep (構成設計)   → { sections[], targetAudience, tone }
  ↓
writeStep (執筆)        → { draft, targetAudience }
  ↓
reviewStep (レビュー)    → { revisions[], article }
```

各ステップが明確な成果物を生成し、次のステップに渡します。

## 誤った分割の例

❌ **悪い例**: 「前半を書く → 後半を書く」
- 前半と後半で文体・深さが揃わない
- 文脈の引き継ぎが不安定

✅ **良い例**: 「research → outline → write → review」
- 各ステップが異なる種類の成果物を生成
- 前ステップの成果が後ステップのインプットになり、品質が積み上がる
